```python
import pandas as pd

from pathlib import Path

# 노트북을 어느 하위 폴더에서 실행해도 저장소 루트를 찾습니다.
_current_dir = Path.cwd().resolve()
PROJECT_ROOT = next(
    path for path in (_current_dir, *_current_dir.parents)
    if (path / "notebooks").is_dir() and (path / "data").is_dir()
)
DATA_DIR = PROJECT_ROOT / "data"
RESULTS_DIR = PROJECT_ROOT / "results"
```

```python
input_dir = RESULTS_DIR / "solana" / "wallet_analysis"
input_paths = [
    input_dir / "sol_top_wallets_all_periods.csv",
    input_dir / "sol_weighted_top_wallets.csv",
]

wallet_frames = []
for input_path in input_paths:
    if not input_path.exists():
        print(f"Wallet file not found: {input_path}")
        continue

    wallets = pd.read_csv(input_path)
    address_column = "address" if "address" in wallets.columns else wallets.columns[0]
    wallet_frames.append(wallets[[address_column]].rename(columns={address_column: "address"}))

if wallet_frames:
    combined_wallets = (
        pd.concat(wallet_frames, ignore_index=True)
        .drop_duplicates()
        .sort_values("address")
        .reset_index(drop=True)
    )
    output_path = input_dir / "sol_combined_top_wallets.csv"
    combined_wallets.to_csv(output_path, index=False)
else:
    print("No wallet files to combine.")
```
