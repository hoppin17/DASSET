파일에는 총 1000개씩 블록이 저장되어 있고, 블록에는 event의 개수에 따라서 크기가 좀 천차만별이다. 여기서 event id값을 우선 뽑아 본 다음에 우리가 필요한 event만 뽑아서 정리해보자. 우리가 원하는 것은 token의 이동과 관련되어 있다. 
- withdraw

- transfer → balances.Transfer, balances.Endowed

- deposit → staking.Rewarded, staking.PayoutStarted

- locked → balances.Locked

이렇게 정리해달라고 하시는데... 뭔가 더있는지 확인해봐야 할 것 같다.

```python
start = 1
while True:
    with open(f'AVAIL_blocks_{start-1}_to_{start+999}.json') as f:
        AVAIL_json = json.load(f)
        temp_df = pd.DataFrame()
    for num in range(999):
        data = AVAIL_json[num]
        events = data['data']['events']
        df = pd.DataFrame(data)
        pd.concat([temp_df, df], ignore_index=True)
    temp_df.to_csv(f'AVAIL_events_{start}_to_{start+999}.csv')
    start += 1000
    # print(json.dumps(AVAIL_json, indent="\t") )
```

```python
start = 1
file_count = 0  # 처리할 총 파일 수를 모르므로 무한 루프로 처리

with tqdm(desc="Overall Progress", unit="file") as pbar:
    while True:
        try:
            with open(f'AVAIL_blocks_{start-1}_to_{start+999}.json') as f:
                AVAIL_json = json.load(f)
            
            temp_df = pd.DataFrame()
            
            for num in tqdm(range(len(AVAIL_json)), desc=f"Processing file {file_count + 1}", leave=False):
                data = AVAIL_json[num]
                timestamp = data['data']['block_timestamp']
                events = data['data']['events']
                df = pd.DataFrame(events)
                df['timestamp'] = timestamp
                temp_df = pd.concat([temp_df, df], ignore_index=True)
            
            temp_df.to_csv(f'../Transaction_events/AVAIL_events_{start}_to_{start+999}.csv', index=False)
            start += 1000
            file_count += 1
            pbar.update(1)
            
        except FileNotFoundError:
            print(f"\nProcessing complete. Total files processed: {file_count}")
            break
        except Exception as e:
            print(f"\nAn error occurred: {str(e)}")
            break
```
