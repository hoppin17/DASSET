```python
import pandas as pd
import os
```

```python
file_list = os.listdir()
```

```python
file_list = [file for file in file_list if 'centrality.csv' in file]
file_list.sort()
file_list
```

```python
temp = pd.read_csv('../9.11_SOL_2020-03-26_2020-12-31_centrality.csv', index_col = 0)
temp
```

```python
pagerank = pd.DataFrame()
betweenness = pd.DataFrame()

for num in range(len(file_list)):
    # 해당 파일을 불러온다
    df = pd.read_csv(f'./Transaction/{file_list[num]}')

    # date를 기준으로 그룹화하고 각 그룹에서 pagerank 기준 상위 15개 선택
    top_15_pagerank = df.groupby('date').apply(lambda x: x.nlargest(15, 'pagerank')).reset_index(drop=True)
    top_15_betweenness = df.groupby('date').apply(lambda x: x.nlargest(15, 'betweenness_centrality')).reset_index(drop=True)

    pagerank = pd.concat([pagerank, top_15_pagerank])
    betweenness = pd.concat([betweenness, top_15_betweenness])
```

```python
pagerank
```

```python
betweenness
```

```python
account_1 = list(pagerank['Unnamed: 0'])
account_2 = list(betweenness['Unnamed: 0'])

account_1.extend(account_2)
account_1
```

```python
A = set(account_1)
A = pd.DataFrame(A)
A
```

```python
A.to_csv('9.17_SOL_2024_top_account')
```
